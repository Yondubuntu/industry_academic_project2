```python
m = folium.Map([lat,long])


for i in df_coin.index[0:10]:
       sub_lat = df_coin.loc [i,"위도"]
       sub_long = df_coin.loc [i,"경도"]
       title = df_coin.loc[i,"상호명"] + " - " + df_coin.loc[i ,"도로명주소"]
       folium.Marker(
       [sub_lat,sub_long] ,
       popup=f'<i>{title}</i>',
       tooltipe = title).add_to(m)

m
```

```python
from folium.plugins import MarkerCluster
```

```python
m = folium.Map([lat,long], zoom_start = 12)
marker_cluster = MarkerCluster().add_to(m)


for i in df_coin.index:
       sub_lat = df_coin.loc [i,"위도"]
       sub_long = df_coin.loc [i,"경도"]
       title = df_coin.loc[i,"상호명"] + " - " + df_coin.loc[i ,"도로명주소"]
       folium.Marker(
       [sub_lat,sub_long] ,
       popup=f'<i>{title}</i>',
       tooltipe = title).add_to(marker_cluster)
m.save("index.html")
m
```
